```python
import pandas

data = pandas.read_csv('Experience-Salary.csv')
features = data.columns

features, data
```

```python
import numpy as np

def reshape(arr):
    return np.array(arr).reshape(-1,1)
```

```python
from sklearn.linear_model import LinearRegression

x, y = reshape(data[features[0]]), reshape(data[features[1]])

reg = LinearRegression(fit_intercept=True)
reg.fit(x, y)
pred = reg.predict(x)

reg.intercept_, reg.coef_, reg.score(x, y), pred
```
